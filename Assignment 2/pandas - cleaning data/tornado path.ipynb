```python
# check if there is any null values in columns
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['storm_date'], format= '%d/%m/%Y')
df
```

```python
df = df[['Date'] + [x for x in df.columns if x != 'Date']]
df
```

```python
df.drop(df.columns[[1]], axis=1, inplace=True)
df
```

```python
df['Year'] = pd.DatetimeIndex(df['Date']).year
df
```

```python
df = df[['Year'] + [x for x in df.columns if x != 'Year']]
df
```

```python
df = df.loc[(df['Year'] >= 2010)]
df
```

```python
df['ID'] = range(1, 1+len(df))
df = df[['ID'] + [x for x in df.columns if x != 'ID']]
df
```
