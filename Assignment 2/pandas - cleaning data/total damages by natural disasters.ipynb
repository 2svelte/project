```python
# import pandas 
import pandas as pd

# import dataset
df = pd.read_csv(r"C:\Users\Nadia\Desktop\UNI\3179\current sem\Assignment 2\datasets\not cleaned yet\natural-disasters (updated).csv")

df.head()
```

```python
df.dropna()
```

```python
# Only keep records of incidents that take place from 2000 above
df = df[df['Year'] >= 1970]
df
```

```python
df = df.sort_values(by='Year')
```

```python
reconstruction = df.groupby('Year')[['Reconstruction costs from earthquakes']].sum().reset_index()
insured = df.groupby('Year')[['Insured damages against earthquakes']].sum().reset_index()
econ = df.groupby('Year')[['Total economic damages from earthquakes']].sum().reset_index()
```

```python
earthquake = reconstruction.set_index('Year').join(insured.set_index('Year'), how = 'inner').join(econ.set_index('Year'), how='inner')
earthquake = earthquake.reset_index()
earthquake
```

```python
reconstruction = df.groupby('Year')[['Number of total people affected by storms']].sum().reset_index()
insured = df.groupby('Year')[['Insured damages against storms']].sum().reset_index()
econ = df.groupby('Year')[['Total economic damages from storms']].sum().reset_index()
```

```python
storms = reconstruction.set_index('Year').join(insured.set_index('Year'), how = 'inner').join(econ.set_index('Year'), how='inner')
storms = storms.reset_index()
storms


```
