```python
# import pandas 
import pandas as pd

# import dataset
df = pd.read_csv(r"C:\Users\Nadia\Desktop\UNI\3179\current sem\Assignment 2\datasets\not cleaned yet\internal_displacement.csv")

df.head()
```

```python
# check if there is any null values in columns
df.isnull().sum()
```

```python
df = df[df['Hazard Type'].notna()]
df
```

```python

df = df.groupby(['Country / Territory', 'Hazard Type'])[['Disaster Internal Displacements']].sum().reset_index()
df
```

```python
df.rename(columns = {'Country / Territory':'Country'}, {'Hazard Type':'Disaster'}, inplace = True)
df
```

```python
df = df[(df['Hazard Type'] == 'Flood') | (df['Hazard Type'] == 'Storm') | (df['Hazard Type'] == 'Earthquake') |
       (df['Hazard Type'] == 'Wildfire') | (df['Hazard Type'] == 'Drought') | (df['Hazard Type'] == 'Volcanic eruption')   ]

df
```

```python
df_to_csv = df.to_csv(r"C:\Users\Nadia\Desktop\UNI\3179\current sem\Assignment 2\datasets\cleaned\internal_displacement(cleaned).csv", index = False)
```
